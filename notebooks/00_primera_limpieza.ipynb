## 0. Primera Limpieza
Primera limpieza para uan exploraci√≥n adecuada

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/seguros.csv')
df = df[['TARGET_FLAG', 'TARGET_AMT','AGE','BLUEBOOK','CAR_AGE','CAR_TYPE','CAR_USE','CLM_FREQ','EDUCATION','HOMEKIDS','HOME_VAL','INCOME','JOB','KIDSDRIV','MSTATUS','MVR_PTS','OLDCLAIM','PARENT1','RED_CAR','REVOKED','SEX','TIF','TRAVTIME','URBANICITY','YOJ']]
money_val = ['BLUEBOOK','HOME_VAL','INCOME','OLDCLAIM']
for i in range(len(money_val)):
    df[money_val[i]] = df[money_val[i]].str.replace('$','')
    df[money_val[i]] =  df[money_val[i]].str.replace(',','')
    df[money_val[i]] = df[money_val[i]].apply(pd.to_numeric)
#CAR_TYPE
df['CAR_TYPE'] = df['CAR_TYPE'].str.replace('z_SUV','SUV')
df['CAR_TYPE'] = df['CAR_TYPE'].str.replace(' ','')
#EDUCATION
df['EDUCATION'] = df['EDUCATION'].str.replace('z_High School','High School')
df['EDUCATION'] = df['EDUCATION'].str.replace(' ','')
#JOB
df['JOB'] = df['JOB'].str.replace('z_Blue Collar','Blue Collar')
df['JOB'] = df['JOB'].str.replace(' ','')
df['JOB'] = df['JOB'].fillna('NotInf')
#MSTATUS
df['MSTATUS'] = df['MSTATUS'].str.replace('z_No','No')
#RED_CAR
df['RED_CAR'] = df['RED_CAR'].str.replace('no','No')
df['RED_CAR'] = df['RED_CAR'].str.replace('yes','Yes')
# SEX
df['SEX'] = df['SEX'].str.replace('z_F','F')
#URBANICITY
df['URBANICITY'] = df['URBANICITY'].str.replace('Highly Urban/ Urban','Urban')
df['URBANICITY'] = df['URBANICITY'].str.replace('z_Highly Rural/ Rural','Rural')
```

```python
file_name = 'seguros_v2.csv'
file_path = '../data/'
df.to_csv(file_path+file_name,index=False)
```
